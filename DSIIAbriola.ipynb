```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Leer el archivo CSV en un DataFrame de pandas
df = pd.read_csv("https://raw.githubusercontent.com/abriolajuan/DSIIAbriola/refs/heads/main/car_price_prediction.csv")

df_precio_autos=df

df_precio_autos.head()
```

```python
pd.reset_option('display.max_rows')
df_precio_autos
```

```python
# Dejar ID como index
df_precio_autos.set_index('ID', inplace=True)
df_precio_autos.head()
```

```python

# Renombramos las columnas
df_precio_autos= df.rename(columns={"ID": "ID", "Price":"PRECIO", "Levy":"IMPUESTO", "Manufacturer":"MARCA", "Model":"MODELO", "Prod. year":"AÑO", "Category":"CATEGORIA", "Leather interior":"CUERO_INTERIOR",
                                 "Fuel type":"TIPO_COMBUSTIBLE", "Engine volume":"VOLUMEN_MOTOR", "Mileage":"KILOMETROS", "Cylinders":"CILINDROS", "Gear box type":"TIPO_CAJA_CAMBIOS",
                                 "Drive wheels":"TRACCION", "Doors":"PUERTAS", "Wheel":"VOLANTE", "Color":"COLOR", "Airbags":"AIRBAGS"})
df_precio_autos.head()
```

```python
#Información sobre las columnas
df_precio_autos.info()
```

```python

# Información sobre los datos de cada columna
resumen=df_precio_autos.describe().round(1)
resumen
```

```python
df_precio_autos.isnull().sum()
```

```python
#Observamos que la tabla kilometros tiene valores como "186005 km", para ello quitamos " km" y convertimos a número
df_precio_autos["KILOMETROS"] = df_precio_autos["KILOMETROS"].str.replace(" km", "").astype(int)
```

```python
df_precio_autos.head()
```

```python
#La columna volumen_motor tambien presenta valores con letras, cuando solo necesitamos el número.
df_precio_autos["VOLUMEN_MOTOR"] = df_precio_autos["VOLUMEN_MOTOR"].str.extract(r"(\d+\.\d+|\d+)").astype(float)
```

```python
df_precio_autos.head()
```

```python
#La columna Puertas presenta valores como "04-May", hay que corregir esto por los valores correctos.
df_precio_autos["PUERTAS"] = df_precio_autos["PUERTAS"].str.extract(r"(\d+)").astype(float)
df_precio_autos.head()
```

```python
#Revisamos si hay valores atípicos, extremadamente altos o valores que sean negativos
plt.figure(figsize=(10,5))
sns.boxplot(x=df_precio_autos["PRECIO"])
plt.title("Distribución de precios")
plt.show()
```

```python
#Eliminamos valores negativos o demasiados altos
df_precio_autos = df_precio_autos[(df_precio_autos["PRECIO"] > 500) & (df_precio_autos["PRECIO"] < 200000)]
```

```python
#Podemos observar que en la columna Impuesto no se registran valores Nulos pero si valores como "-"
df_precio_autos["IMPUESTO"].unique()
```

```python
df_precio_autos["IMPUESTO"] = df_precio_autos["IMPUESTO"].replace("-", "0")  # Reemplazar "-" con 0
df_precio_autos["IMPUESTO"] = pd.to_numeric(df_precio_autos["IMPUESTO"])  # Convertir a número
```

```python
df_precio_autos["IMPUESTO"] = pd.to_numeric(df_precio_autos["IMPUESTO"])
```

```python
df_precio_autos["IMPUESTO"].isna().sum()
```

```python
#Visualizamos los valores para ver si hay valores atípicos
plt.figure(figsize=(10,5))
sns.boxplot(x=df_precio_autos["IMPUESTO"])
plt.title("Distribución del Impuesto")
plt.show()
```

```python
#Definimos un rango lógico y filtramos
df_precio_autos = df_precio_autos[(df_precio_autos["IMPUESTO"] >= 0) & (df_precio_autos["IMPUESTO"] < 10000)]
```

```python
#Por ultimo analizamos la columna Precio
print(df_precio_autos["PRECIO"].dtype)
print(df_precio_autos["PRECIO"].unique()[:20])  # Muestra los primeros 20 valores únicos
```

```python
print(df_precio_autos["PRECIO"].isna().sum())  # Cantidad de valores NaN
```

```python
print(df_precio_autos["PRECIO"].describe().round(2))
```

```python
#Visualizamos la distribucion de Precios
sns.histplot(df_precio_autos["PRECIO"], bins=50, kde=True)
plt.title("Distribución de Precios de Autos")
plt.xlabel("Precio")
plt.show()
```

```python
sns.boxplot(x=df_precio_autos["PRECIO"])
plt.title("Boxplot de Precios de Autos")
plt.show()
```

Observando y analizando el boxplot, no podemos definir con claridad si hay valores atípicos ya que no hay ninguno que sea demasiado alto con respecto a los demas. Tampoco observamos valores nulos ni negativos.

```python
#Realizamos Histogramas de Precios y Kilómetros, ademas un TOP 10 de marcas

# Histograma de precios
sns.histplot(df_precio_autos["PRECIO"], bins=50, kde=True)
plt.title("Distribución de Precios de Autos")
plt.xlabel("Precio")
plt.show()

# Histograma de kilómetros
sns.histplot(df_precio_autos["KILOMETROS"], bins=50, kde=True)
plt.title("Distribución de Kilómetros")
plt.xlabel("Kilómetros")
plt.show()

# Top 10 marcas más vendidas
df_precio_autos["MARCA"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Marcas Más Comunes")
plt.ylabel("Cantidad de Autos")
plt.show()

```

```python
#Ahora realizamos un análisis bivariado.
# Scatterplot Precio vs Kilómetros
sns.scatterplot(x=df_precio_autos["KILOMETROS"], y=df_precio_autos["PRECIO"])
plt.title("Relación entre Precio y Kilómetros")
plt.xlabel("Kilómetros")
plt.ylabel("Precio")
plt.show()
```

```python
# Boxplot de precios por marca (Top 10 marcas más comunes)
top_marcas = df_precio_autos["MARCA"].value_counts().head(10).index
sns.boxplot(x=df_precio_autos[df_precio_autos["MARCA"].isin(top_marcas)]["MARCA"],
            y=df_precio_autos[df_precio_autos["MARCA"].isin(top_marcas)]["PRECIO"])
plt.xticks(rotation=45)
plt.title("Distribución de Precios por Marca")
plt.show()
```

```python
# Filtrar solo columnas numéricas
df_numerico = df_precio_autos.select_dtypes(include=["number"])

# Matriz de correlación
plt.figure(figsize=(10,6))
sns.heatmap(df_numerico.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlación")
plt.show()
```

# Principales observaciones:

***Correlaciones fuertes:***
- A mayor volumen del motor, más cilindros tiene el auto. (0.77)

Los motores con mayor volumen suelen tener más cilindros.

- Autos mas nuevos tienden a pagar mas impuestos. (0.40)

En muchos estados de EE.UU., los impuestos de registro o matrícula pueden aumentar con el valor del vehículo. Los autos nuevos, por lo general, tienen un valor más alto, lo que puede resultar en un costo de matrícula más elevado.

- Los autos con motores más grandes pagan mas impuestos. (0.34)

Algunos estados y condados imponen impuestos o tarifas más altas a los vehículos menos eficientes en combustible o con motores más grandes.

***Correlaciones moderadas:***

-Los autos mas nuevos suelen costar más. (0.33)

Los autos nuevos suelen costar mas debido a los avances tecnológicos, valor de reventa, mejoras en eficiencia y rendimiento.

-Autos con motores mas grandes tienden a ser mas caros. (0.16)

Si bien no observamos una alta correlacion, el volumen del motor impacta sobre el precio de un auto.

-La cantidad de cilindros impacta sobre el precio.(0.13)

Podemos observar un leve impacto en el precio de un vehiculo segun la cantidad de cilindros.

***Correlaciones nulas:***

-Kilometros. (-0.02)

No parece influir mucho en el precio, lo cual parece raro ya que suele ser que a mayor cantidad de kilometros el precio disminuye. Habria que realizar un analisis mas exaustivo sobre los datos de los kilometros registrados...

No impactan en el precio. Esto puede ser correcto.
