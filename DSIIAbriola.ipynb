```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Leer el archivo CSV en un DataFrame de pandas
df = pd.read_csv("https://raw.githubusercontent.com/abriolajuan/DSIIAbriola/refs/heads/main/car_price_prediction.csv")

df_precio_autos=df

df_precio_autos.head()
```

```python
pd.reset_option('display.max_rows')
df_precio_autos
```

```python
# Dejar ID como index
df_precio_autos.set_index('ID', inplace=True)
df_precio_autos.head()
```

```python

# Renombramos las columnas
df_precio_autos= df.rename(columns={"ID": "ID", "Price":"PRECIO", "Levy":"IMPUESTO", "Manufacturer":"MARCA", "Model":"MODELO", "Prod. year":"AÃ‘O", "Category":"CATEGORIA", "Leather interior":"CUERO_INTERIOR",
                                 "Fuel type":"TIPO_COMBUSTIBLE", "Engine volume":"VOLUMEN_MOTOR", "Mileage":"KILOMETROS", "Cylinders":"CILINDROS", "Gear box type":"TIPO_CAJA_CAMBIOS",
                                 "Drive wheels":"TRACCION", "Doors":"PUERTAS", "Wheel":"VOLANTE", "Color":"COLOR", "Airbags":"AIRBAGS"})
df_precio_autos.head()
```

```python
#InformaciÃ³n sobre las columnas
df_precio_autos.info()
```

```python

# InformaciÃ³n sobre los datos de cada columna
resumen=df_precio_autos.describe().round(1)
resumen
```

```python
df_precio_autos.isnull().sum()
```

```python
#Observamos que la tabla kilometros tiene valores como "186005 km", para ello quitamos " km" y convertimos a nÃºmero
df_precio_autos["KILOMETROS"] = df_precio_autos["KILOMETROS"].str.replace(" km", "").astype(int)
```

```python
df_precio_autos.head()
```

```python
#La columna volumen_motor tambien presenta valores con letras, cuando solo necesitamos el nÃºmero.
df_precio_autos["VOLUMEN_MOTOR"] = df_precio_autos["VOLUMEN_MOTOR"].str.extract(r"(\d+\.\d+|\d+)").astype(float)
```

```python
df_precio_autos.head()
```

```python
#La columna Puertas presenta valores como "04-May", hay que corregir esto por los valores correctos.
df_precio_autos["PUERTAS"] = df_precio_autos["PUERTAS"].str.extract(r"(\d+)").astype(float)
df_precio_autos.head()
```

```python
#Revisamos si hay valores atÃ­picos, extremadamente altos o valores que sean negativos
plt.figure(figsize=(10,5))
sns.boxplot(x=df_precio_autos["PRECIO"])
plt.title("DistribuciÃ³n de precios")
plt.show()
```

```python
#Eliminamos valores negativos o demasiados altos
df_precio_autos = df_precio_autos[(df_precio_autos["PRECIO"] > 500) & (df_precio_autos["PRECIO"] < 200000)]
```

```python
#Podemos observar que en la columna Impuesto no se registran valores Nulos pero si valores como "-"
df_precio_autos["IMPUESTO"].unique()
```

```python
df_precio_autos["IMPUESTO"] = df_precio_autos["IMPUESTO"].replace("-", "0")  # Reemplazar "-" con 0
df_precio_autos["IMPUESTO"] = pd.to_numeric(df_precio_autos["IMPUESTO"])  # Convertir a nÃºmero
```

```python
df_precio_autos["IMPUESTO"] = pd.to_numeric(df_precio_autos["IMPUESTO"])
```

```python
df_precio_autos["IMPUESTO"].isna().sum()
```

```python
#Visualizamos los valores para ver si hay valores atÃ­picos
plt.figure(figsize=(10,5))
sns.boxplot(x=df_precio_autos["IMPUESTO"])
plt.title("DistribuciÃ³n del Impuesto")
plt.show()
```

```python
#Definimos un rango lÃ³gico y filtramos
df_precio_autos = df_precio_autos[(df_precio_autos["IMPUESTO"] >= 0) & (df_precio_autos["IMPUESTO"] < 10000)]
```

```python
#Por ultimo analizamos la columna Precio
print(df_precio_autos["PRECIO"].dtype)
print(df_precio_autos["PRECIO"].unique()[:20])  # Muestra los primeros 20 valores Ãºnicos
```

```python
print(df_precio_autos["PRECIO"].isna().sum())  # Cantidad de valores NaN
```

```python
print(df_precio_autos["PRECIO"].describe().round(2))
```

```python
#Visualizamos la distribucion de Precios
sns.histplot(df_precio_autos["PRECIO"], bins=50, kde=True)
plt.title("DistribuciÃ³n de Precios de Autos")
plt.xlabel("Precio")
plt.show()
```

```python
sns.boxplot(x=df_precio_autos["PRECIO"])
plt.title("Boxplot de Precios de Autos")
plt.show()
```

```python
#Realizamos Histogramas de Precios y KilÃ³metros, ademas un TOP 10 de marcas

# Histograma de precios
sns.histplot(df_precio_autos["PRECIO"], bins=50, kde=True)
plt.title("DistribuciÃ³n de Precios de Autos")
plt.xlabel("Precio")
plt.show()

# Histograma de kilÃ³metros
sns.histplot(df_precio_autos["KILOMETROS"], bins=50, kde=True)
plt.title("DistribuciÃ³n de KilÃ³metros")
plt.xlabel("KilÃ³metros")
plt.show()

# Top 10 marcas mÃ¡s vendidas
df_precio_autos["MARCA"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Marcas MÃ¡s Comunes")
plt.ylabel("Cantidad de Autos")
plt.show()

```

```python
#Ahora realizamos un anÃ¡lisis bivariado.
# Scatterplot Precio vs KilÃ³metros
sns.scatterplot(x=df_precio_autos["KILOMETROS"], y=df_precio_autos["PRECIO"])
plt.title("RelaciÃ³n entre Precio y KilÃ³metros")
plt.xlabel("KilÃ³metros")
plt.ylabel("Precio")
plt.show()
```

```python
# Boxplot de precios por marca (Top 10 marcas mÃ¡s comunes)
top_marcas = df_precio_autos["MARCA"].value_counts().head(10).index
sns.boxplot(x=df_precio_autos[df_precio_autos["MARCA"].isin(top_marcas)]["MARCA"],
            y=df_precio_autos[df_precio_autos["MARCA"].isin(top_marcas)]["PRECIO"])
plt.xticks(rotation=45)
plt.title("DistribuciÃ³n de Precios por Marca")
plt.show()
```

```python
# Filtrar solo columnas numÃ©ricas
df_numerico = df_precio_autos.select_dtypes(include=["number"])

# Matriz de correlaciÃ³n
plt.figure(figsize=(10,6))
sns.heatmap(df_numerico.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de CorrelaciÃ³n")
plt.show()
```
