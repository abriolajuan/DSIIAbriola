Resumen
En este proyecto se lleva a cabo un análisis de un conjunto de datos sobre vehículos con el objetivo de entender los factores que influyen en el precio de un auto usado. El dataset contiene variables clave como precio, marca, modelo, kilometraje, cantidad de cilindros, antigüedad del vehículo, entre otros. A través del análisis de estos datos, se busca identificar las características más relevantes que afectan el precio de un vehículo, con miras a predecir dicho precio en el futuro, en función de estas variables.

Objetivo
El principal objetivo de este análisis es identificar los factores más relevantes que influyen en el precio de un vehículo usado. Esto permitirá desarrollar un modelo que prediga el precio de un auto basándose en sus características, facilitando así el proceso de tasación o compra/venta en mercados de vehículos.

Preguntas de Investigación
Para guiar el análisis, se plantearon las siguientes preguntas:

¿Qué factores influyen más en el precio de un auto usado?

Esta pregunta busca identificar qué variables tienen un mayor impacto en la determinación del precio de un vehículo.
¿Cómo afecta la antigüedad y el kilometraje al valor de un auto?

Se plantea investigar la relación entre la antigüedad del vehículo y su kilometraje con respecto al precio. ¿A medida que un auto envejece y acumula más kilómetros, su precio disminuye?
¿Influye la marca y la categoría del vehículo en su precio?

En este análisis se evalúa si existen diferencias significativas en el precio dependiendo de la marca del vehículo y su categoría (por ejemplo, sedán, SUV, etc.).
¿Qué tanto influye en el precio el volumen del motor y la cantidad de cilindros?

La potencia del motor y el número de cilindros pueden afectar el precio de un vehículo. Este análisis busca cuantificar esa relación.
¿El color y tipo de combustible tienen algún impacto sobre el precio?

Algunas veces, las preferencias de los consumidores pueden influir en el valor de un vehículo. En este caso, se evalúa si el color del vehículo o el tipo de combustible utilizado tienen algún efecto sobre su precio.

```python
# Seleccionamos las 8 marcas más frecuentes
top_marca = df_precio_autos["MARCA"].value_counts().nlargest(8).index

# Seleccionamos los 10 colores más frecuentes
top_color = df_precio_autos["COLOR"].value_counts().nlargest(10).index

# Filtramos por marcas y colores más frecuentes
df_precio_autos_filtered = df_precio_autos[df_precio_autos["MARCA"].isin(top_marca) & df_precio_autos["COLOR"].isin(top_color)]

# Definimos las variables categóricas
categorical_vars = ["MARCA", "CATEGORIA", "TIPO_COMBUSTIBLE", "TRACCION", "COLOR"]

# Creamos los gráficos con los datos filtrados
fig, axes = plt.subplots(1, len(categorical_vars), figsize=(20, 5))
for i, var in enumerate(categorical_vars):
    sns.boxplot(x=df_precio_autos_filtered[var], y=df_precio_autos_filtered["PRECIO"], ax=axes[i])
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=45, fontsize=8)
    axes[i].set_title(f"Distribución de PRECIO según {var}")
plt.tight_layout()
plt.show()

```

```python
import scipy.stats as stats

for var in categorical_vars:
    grupos = [df_precio_autos[df_precio_autos[var] == cat]["PRECIO"] for cat in df_precio_autos[var].unique()]
    anova_result = stats.f_oneway(*grupos)
    print(f"ANOVA para {var}: p-valor = {anova_result.pvalue}")
```

Según el análisis de varianza realizado, podemos observar que todas las variables tienen un p-valor extremadamente bajo o nulo. Esto significa que rechazamos la hipótesis nula, lo que indica que cada una de estas variables tiene un impacto significativo sobre la variable Precio.

```python
#Feature Engneering
#Creamos nuevas columnas para analizar si se mejora la correlación con el precio

#ANTIGUEDAD DEL AUTO
df_precio_autos["ANTIGUEDAD"] = 2025 - df_precio_autos["AÑO"]

#RANGOS DE PRECIOS
df_precio_autos["RANGO_PRECIO"] = pd.cut(df_precio_autos["PRECIO"], bins=[0, 5000, 15000, 30000, 50000, 100000, 200000], labels=['Muy Bajo', 'Bajo', 'Medio', 'Alto', 'Muy Alto', 'Súper Alto'])

#RELACION KMS / AÑOS
df_precio_autos["KM_ANIO"] = df_precio_autos["KILOMETROS"] / df_precio_autos["ANTIGUEDAD"]

#RELACION PRECIO / KMS
df_precio_autos["PRECIO_KM"] = df_precio_autos["PRECIO"] / df_precio_autos["KILOMETROS"]

df_precio_autos.head()



```

```python
variables_interes = ["ANTIGUEDAD", "KM_ANIO", "PRECIO_KM", "PRECIO"]
df_interes = df_precio_autos[variables_interes]

# Correlación de las variables creadas con el precio
correlacion = df_interes.corr()

# Visualizamos la matriz de correlación
plt.figure(figsize=(6, 4))
sns.heatmap(correlacion, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlación de las variables creadas con el Precio")
plt.show()
```

Segun lo observado la correlación entre antiguedad y precio del auto sugiere que los autos mas antiguos tienden a tener un precio mas bajo.

```python
# Scatterplot de Antigüedad vs Precio
sns.scatterplot(x=df_precio_autos["ANTIGUEDAD"], y=df_precio_autos["PRECIO"])
plt.title("Relación entre Antigüedad y Precio")
plt.xlabel("Antigüedad (Años)")
plt.ylabel("Precio")
plt.show()
```

Segun podemos observar en el grafico, se puede decir que a mayor antiguedad menor es el precio. Si bien esto no esta claro si se puede utilizar como una afirmación.
